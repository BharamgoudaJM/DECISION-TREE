```python
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
from sklearn import preprocessing
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import metrics

```

```python
mj = pd.read_csv("Fraud_check.csv")
```

```python
mj
```

```python
mj.insert(6,"tax.cond",'')
```

```python
mj
```

```python
for i in range(0, len(mj['Taxable.Income'])):
    if mj['Taxable.Income'][i] <= 30000:
        mj['tax.cond'][i] = 'Risky'
    else:
        mj['tax.cond'][i] = 'Good'
```

```python
mj
```

```python
mj.info()
```

```python
mj1=preprocessing.LabelEncoder()
```

```python
objlist = ['Undergrad','Marital.Status','Urban','tax.cond']
mj[objlist] = mj[objlist].apply(mj1.fit_transform)
```

```python
mj.columns
```

```python
mj2 = mj[['Undergrad','Marital.Status','City.Population','Work.Experience','Urban','tax.cond']]
```

```python
mj2
```

```python
x = mj2.iloc[:,:-1]
y = mj2.iloc[:,-1]
y
```

# Building Decision Tree CLassifier Using Gini Criteria

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
```

```python
model = DecisionTreeClassifier(criterion='gini',min_samples_split=5,max_depth=10)
```

```python
model.fit(x_train,y_train)
pred = model.predict(x_train)
metrics.accuracy_score(pred,y_train)
```

```python
model.fit(x_train,y_train)
pred = model.predict(x_test)
pd.Series(pred).value_counts()
```

```python
metrics.accuracy_score(pred,y_test)
```

```python
print(classification_report(y_test,pred))
```

# Building Decision Tree Classifier using Entropy Criteria

```python
model = DecisionTreeClassifier(criterion='entropy',min_samples_split=5)
```

```python
model.fit(x_train,y_train)
pred = model.predict(x_test)
metrics.accuracy_score(pred,y_test)
```

```python
print(classification_report(y_test,pred))
```

# BAGGING

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import BaggingClassifier
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)

num_trees = 100
model = BaggingClassifier(max_samples=0.8, n_estimators=num_trees,random_state=8)

model.fit(x_train,y_train)
pred = model.predict(x_test)
print(classification_report(y_test,pred))
```
